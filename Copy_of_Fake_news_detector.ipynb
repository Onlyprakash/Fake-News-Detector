```python
import pandas as pd

fake = pd.read_csv(
    'data/Fake.csv',
    engine='python',
    sep=',',
    quoting=3,
    on_bad_lines='skip'
)

true = pd.read_csv(
    'data/True.csv',
    engine='python',
    sep=',',
    quoting=3,
    on_bad_lines='skip'
)

print("Fake shape:", fake.shape)
print("True shape:", true.shape)
fake.head()
```

```python
# 0 = Fake news
fake['label'] = 0

# 1 = Real news
true['label'] = 1
data = pd.concat([fake, true])
data = data.sample(frac=1, random_state=42).reset_index(drop=True)
data.head()
data['label'].value_counts()


```

The dataset was labeled and shuffled to ensure unbiased learning during model training.

```python
import re

def clean_text(text):
    text = text.lower()                      # lowercase
    text = re.sub(r'\W', ' ', text)          # remove special characters
    text = re.sub(r'\s+', ' ', text)         # remove extra spaces
    return text
    data['text'] = data['text'].apply(clean_text)


```

```python
data['text'].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data = data.dropna(subset=['text'])
data = data.reset_index(drop=True)
```

```python
X = data['text']
y = data['label']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    stop_words='english',
    max_df=0.7
)

X_tfidf = tfidf.fit_transform(X)
X_tfidf.shape

```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
def predict_news(news_text):
    cleaned_text = clean_text(news_text)
    vector = tfidf.transform([cleaned_text])
    prediction = model.predict(vector)

    if prediction[0] == 1:
        return "REAL NEWS"
    else:
        return "FAKE NEWS"
```

```python
predict_news("Aliens landed on earth and are controlling world leaders")
```

```python
predict_news("Supreme Court issues new guidelines on digital privacy laws")
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train, y_train)
```

```python
y_pred_lr = lr_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_lr = accuracy_score(y_test, y_pred_lr)
print("Logistic Regression Accuracy:", accuracy_lr)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_lr))
```

```python
def predict_news_lr(news_text):
    cleaned_text = clean_text(news_text)
    vector = tfidf.transform([cleaned_text])
    prediction = lr_model.predict(vector)

    return "REAL NEWS" if prediction[0] == 1 else "FAKE NEWS"
```
